```python
# Plot how data field values are distributed across the locations 'Argentina-Buenos_Aires', 'Argentina-CABA', 'Argentina-Catamarca'


import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("data.csv")

#Droping null values and modifying the data
df.drop(['time_period','time_period_type'],axis=1,inplace=True)
df['report_date']=df['report_date'].str.replace('_','-',regex=False)

#Retrieving the required data
result_argen_catamarca=df.query('location=="Argentina-Catamarca" and value !=0')[["value","data_field"]]
result_argen_caba=df.query('location=="Argentina-CABA" and value !=0')[["value","data_field"]]
result_argen_Buenos_Aires=df.query('location=="Argentina-Buenos_Aires" and value !=0')[["value","data_field"]]

#Modifying the data for ploting 
result_argen_catamarca['value']=pd.to_numeric(result_argen_catamarca['value'])
result_argen_caba['value']=pd.to_numeric(result_argen_caba['value'])
result_argen_Buenos_Aires['value']=pd.to_numeric(result_argen_Buenos_Aires['value'])

#Grouping the data according to the "data_field"
result_argen_catamarca=result_argen_catamarca.groupby("data_field").sum()
result_argen_caba=result_argen_caba.groupby("data_field").sum()
result_argen_Buenos_Aires=result_argen_Buenos_Aires.groupby("data_field").sum()

#Plotting the graph
result_argen_catamarca.plot(kind='pie',autopct='%.2f',y="value",figsize=(10,6),title="Argentina Catamarca",explode=[0.1,0,0,0,0,0])
plt.legend( loc = 'upper left', frameon=False)
result_argen_caba.plot(kind='pie',autopct='%.2f',y="value",explode=[0.4,0.1,0,0.2,0,0.6],title="Argentina-CABA",figsize=(10,6))
plt.legend( loc = 'upper left', frameon=False)
result_argen_Buenos_Aires.plot(kind='pie',autopct='%.2f',explode=[0.1,0,0.2,0,0.4,0.6],title="Argentina-Buenos_Aires",y='value',figsize=(10,6))
plt.legend( loc = 'upper left', frameon=False)
                            


```
