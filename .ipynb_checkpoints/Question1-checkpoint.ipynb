```python
#show data field code 'AR0005' values for 'Argentina-Catamarca' for entire duration
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("data.csv")

df.drop(['time_period','time_period_type'],axis=1,inplace=True)
df['report_date']=df['report_date'].str.replace('_','-',regex=False)
df.groupby("report_date").count()

#retrieving required data
ploting=df.query('data_field_code=="AR0005" & location=="Argentina-Catamarca"')[["report_date","value"]]

#Modifying the data 
ploting['value']=pd.to_numeric(ploting['value'])
#Plotting the data
result=ploting.plot(kind='line',x='report_date',y='value',figsize=(10,5),title="Plot of values for a loc",linestyle='-',marker='o',grid=True)
plt.xlabel("reporting date")
plt.ylabel("values")
```
