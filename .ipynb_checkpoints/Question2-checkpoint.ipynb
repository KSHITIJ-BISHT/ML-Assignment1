```python
#Plot 'Argentina-CABA' and 'Argentina-Catamarca' for number of cases values for entire duration to identify the difference.

import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("data.csv")
df.drop(['time_period','time_period_type'],axis=1,inplace=True)
df['report_date']=df['report_date'].str.replace('_','-',regex=False)

result_argen_catamarca=df.query('location=="Argentina-Catamarca" and value !=0')[["value","report_date"]]
result_argen_caba=df.query('location=="Argentina-CABA" and value !=0')[["value","report_date"]]

result_argen_catamarca['value']=pd.to_numeric(result_argen_catamarca['value'])
plot_catamarca=result_argen_catamarca.groupby("report_date").sum()
result_argen_caba['value']=pd.to_numeric(result_argen_caba['value'])
plot_caba=result_argen_caba.groupby("report_date").sum()

result= pd.merge(plot_catamarca,plot_caba,on=["report_date"])
annotate=result.plot(kind="bar",figsize=(15,6),grid=True,title="Comaprison of two locations for number of cases")
plt.legend(["Argentina-Catamarca", "Argentina-CABA"]);
for p in annotate.patches:
    plt.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.ylabel("VALUES")
plt.xlabel("REPORTING DATE")
```
